```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
xpoints = np.array([0,6])
ypoints = np.array([0,250])
```

```python
plt.plot(xpoints,ypoints)
plt.show()
```

```python
plt.plot(xpoints,ypoints,marker='o')
plt.show()
```

```python
ypoints = np.array([8,5,94])
plt.plot(ypoints,marker='*',linestyle='dotted')
plt.show()
```

```python
xpoints = np.array([1,2,3,4,5])
ypoints = np.array([2,5,3,6,4])
plt.plot(xpoints,ypoints,marker='*',linestyle='dashed')
plt.xlabel("Average")
plt.ylabel("Calories")
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,6])
plt.subplot(2,2,1)
plt.plot(x,y)
m=np.array([0,1,2,3])
n=np.array([10,20,30,40])
plt.subplot(2,2,2)
plt.plot(m,n)
plt.show()
```

```python
x = np.array([1,3,4,2,5,7,6,8])
y = np.array([78,120,89,32,16,67,150,90])
plt.scatter(x,y)
plt.show()
```

```python
Categories = np.array(['Category A', 'Category B', 'Category C', 'Category D', 'Category E'])
Values = np.array ([23, 56, 41, 62, 19])
plt.subplot(1,1,1)
plt.bar(Categories,Values)
plt.show()
```

```python
Labels = np.array(['Apple', 'Banana', 'Orange', 'Mango'])
Sizes = np.array ([30, 25, 15, 30])
plt.pie(Sizes)
plt.show()
```
