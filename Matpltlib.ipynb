```python
import matplotlib.pyplot as plt 
import numpy as np
```

In the first subplot (top-left), plot a bar chart using the following data:
Categories: ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
Values: [23, 56, 41, 62, 19]

```python
Categories = np.array(['Cat A', 'Cat B', 'Cat C', 'Cat D', 'Cat E'])
Values = np.array ([23, 56, 41, 62, 19])
plt.subplot(2,1,1)
plt.bar(Categories,Values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

In the second subplot (top-right), plot a histogram using the following data:
Data: [12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10]

```python
Data = np.array ([12, 17, 21, 18, 14, 13, 16, 9, 12, 15, 19, 11, 14, 16, 20, 18, 15, 13, 16, 11, 10])
plt.subplot(2,2,2)
plt.hist(Data)
plt.title('Histogram of Data')
plt.xlabel("Data")
plt.ylabel("Values")
plt.show()
```

In the third subplot (bottom-left), plot a pie chart using the following data:
Labels: ['Apple', 'Banana', 'Orange', 'Mango']
Sizes: [30, 25, 15, 30]

```python
Labels = np.array(['Apple', 'Banana', 'Orange', 'Mango'])
Sizes = np.array ([30, 25, 15, 30])
plt.subplot(2,2,3)
plt.pie(Sizes, labels = Labels)
plt.show()
```

In the fourth subplot (bottom-right), plot a scatter plot using the following data:
X values: [1, 2, 3, 4, 5]
Y values: [2, 5, 3, 6, 4]

```python
x = np.array([1,2,3,4,5])
y = np.array([2,5,3,6,4])
plt.subplot(2,2,4)
plt.scatter(x,y)
plt.show()
```
